```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities_data_path = os.path.join('cities.csv')
cities_df = pd.read_csv(cities_data_path)
```

```python
cities_df.head()
```

```python
cities_df['City'].nunique()
```

```python
clean_df = cities_df.drop_duplicates(subset='City')
```

```python
clean_df['City'].count()
```

```python
cities_data_df = clean_df.reset_index(drop=True)
cities_data_df.columns
```

```python
cities_data_df.to_html('cities_data.html')
```

```python
lat = cities_data_df['Lat'].values
lon = cities_data_df['Lng'].values
hum = cities_data_df['Humidity'].values
clo = cities_data_df['Cloudiness'].values
win = cities_data_df['Wind Speed'].values
mxtem = cities_data_df['Max Temp'].values
```

```python
# Lon vs Lat Plt
plt.scatter(lon, lat, marker="o", facecolors="red", edgecolors="black",)
plt.title('City Longitude vs Latitude')
plt.xlabel('Longitude')
plt.ylabel('City Latitude')
plt.grid()

# Save the Figure
plt.savefig("Images/longitude_vs_latitude.png")
plt.show()
```

```python
# Lat vs Max Temp Plt
plt.scatter(lat, mxtem, marker="o", facecolors="red", edgecolors="black",)
plt.title('City Latitude vs Max Temperature')
plt.xlabel('Latitude')
plt.ylabel('City Max Temperature (F)')
plt.grid()

# Save the Figure
plt.savefig("Images/latitude_vs_max_temp.png")
plt.show()
```

```python
# Lat vs humidity Plt
plt.scatter(lat, hum, marker="o", facecolors="red", edgecolors="black",)
plt.title('City Latitude vs Humidity')
plt.xlabel('Latitude')
plt.ylabel('Humidity(%)')
plt.grid()

# Save the Figure
plt.savefig("Images/latitude_vs_humidity.png")
plt.show()
```

```python
# Lat vs cloudiness Plt
plt.scatter(lat, clo, marker="o", facecolors="red", edgecolors="black",)
plt.title('City Latitude vs Cloudiness')
plt.xlabel('Latitude')
plt.ylabel('Clodiness(%)')
plt.grid()

# Save the Figure
plt.savefig("Images/latitude_vs_cloudiness.png")
plt.show()
```

```python
# Lat vs temperature Plt
plt.scatter(lat, win, marker="o", facecolors="red", edgecolors="black",)
plt.title('City Latitude vs Wind Speed')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (MPH)')
plt.grid()

# Save the Figure
plt.savefig("Images/latitude_vs_win_speed.png")
plt.show()
```
